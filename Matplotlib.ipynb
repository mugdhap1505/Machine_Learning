```python
df1 = np.arange(4)
plt.bar(df1 + 0.00, df1[1], color = 'y', width = 0.25)
plt.bar(df1 + 0.25, df1[2], color = 'g', width = 0.25)
plt.bar(df1 + 0.50, df1[3], color = 'c', width = 0.25)
plt.show()
```
